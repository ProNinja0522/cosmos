# Dimensionality Reduction on Digits Dataset

This notebook demonstrates how six different dimensionality reduction techniques reveal structure in the high-dimensional `digits` dataset.

The dataset consists of 8×8 grayscale images of handwritten digits (0–9), with each image flattened into a 64-dimensional vector.

**Included Methods:**
1. PCA
2. MDS
3. Isomap
4. Laplacian Eigenmap
5. t-SNE
6. Autoencoder

```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load digits dataset
digits = load_digits()
X = digits.data
y = digits.target

# Display first 20 digit images
fig, axes = plt.subplots(2, 10, figsize=(12, 3))
for i, ax in enumerate(axes.ravel()):
    ax.imshow(digits.images[i], cmap='gray')
    ax.set_title(f'Label: {digits.target[i]}')
    ax.axis('off')
plt.suptitle('Example Digits (8×8 images)')
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
X_scaled = StandardScaler().fit_transform(X)
```

## PCA

This method reduces 64-dimensional image vectors to 2D for visualization.

```python
from sklearn.decomposition import PCA
X_pca = PCA(n_components=2).fit_transform(X_scaled)
```

```python
plt.figure(figsize=(6, 5))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='tab10', s=10)
plt.colorbar(label='Digit Label')
plt.title('PCA Projection of Digits')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.grid(True)
plt.show()
```

## MDS

This method reduces 64-dimensional image vectors to 2D for visualization.

```python
from sklearn.manifold import MDS
X_mds = MDS(n_components=2, random_state=42).fit_transform(X_scaled)
```

```python
plt.figure(figsize=(6, 5))
plt.scatter(X_mds[:, 0], X_mds[:, 1], c=y, cmap='tab10', s=10)
plt.colorbar(label='Digit Label')
plt.title('MDS Projection of Digits')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.grid(True)
plt.show()
```

## Isomap

This method reduces 64-dimensional image vectors to 2D for visualization.

```python
from sklearn.manifold import Isomap
X_iso = Isomap(n_components=2).fit_transform(X_scaled)
```

```python
plt.figure(figsize=(6, 5))
plt.scatter(X_iso[:, 0], X_iso[:, 1], c=y, cmap='tab10', s=10)
plt.colorbar(label='Digit Label')
plt.title('Isomap Projection of Digits')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.grid(True)
plt.show()
```

## Laplacian Eigenmap

This method reduces 64-dimensional image vectors to 2D for visualization.

```python
from sklearn.manifold import SpectralEmbedding
X_lap = SpectralEmbedding(n_components=2, random_state=42).fit_transform(X_scaled)
```

```python
plt.figure(figsize=(6, 5))
plt.scatter(X_lap[:, 0], X_lap[:, 1], c=y, cmap='tab10', s=10)
plt.colorbar(label='Digit Label')
plt.title('Laplacian Eigenmap Projection of Digits')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.grid(True)
plt.show()
```

## t-SNE

This method reduces 64-dimensional image vectors to 2D for visualization.

```python
from sklearn.manifold import TSNE
X_tsne = TSNE(n_components=2, random_state=42).fit_transform(X_scaled)
```

```python
plt.figure(figsize=(6, 5))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='tab10', s=10)
plt.colorbar(label='Digit Label')
plt.title('t-SNE Projection of Digits')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.grid(True)
plt.show()
```

## Autoencoder

This method reduces 64-dimensional image vectors to 2D for visualization.

```python
from sklearn.neural_network import MLPRegressor
ae = MLPRegressor(hidden_layer_sizes=(32,), max_iter=2000, random_state=42)
ae.fit(X_scaled, X_scaled)
X_auto = ae.predict(X_scaled)[:, :2]
```

```python
plt.figure(figsize=(6, 5))
plt.scatter(X_auto[:, 0], X_auto[:, 1], c=y, cmap='tab10', s=10)
plt.colorbar(label='Digit Label')
plt.title('Autoencoder Projection of Digits')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.grid(True)
plt.show()
```
