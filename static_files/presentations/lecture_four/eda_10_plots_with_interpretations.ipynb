# EDA: 10 Bivariate and Univariate Plotting Methods with Interpretations

## 1. Scatter Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
df_titanic = sns.load_dataset('titanic').dropna(subset=['age', 'fare'])
sns.scatterplot(x='age', y='fare', data=df_titanic)
plt.title('Titanic: Age vs. Fare (Scatter Plot)')
plt.show()
```

### 🔍 Interpretation:
- **Upward or downward trend**: Suggests correlation between variables.
- **Tight cluster**: Strong relationship or consistent behavior.
- **Outliers**: May suggest unusual observations or data errors.


## 2. Joint Plot

```python
sns.jointplot(x='age', y='fare', data=df_titanic, kind='scatter')
plt.suptitle('Titanic: Age vs. Fare (Joint Plot)', y=1.02)
plt.show()
```

### 🔍 Interpretation:
- **Center clustering**: Shows where most data lie.
- **Marginal histograms**: Help understand distribution of each variable.
- **If using KDE or hex**, highlights concentration zones.


## 3. LM Plot (Linear Regression)

```python
sns.lmplot(x='age', y='fare', data=df_titanic, height=6)
plt.title('Titanic: Age vs. Fare (LM Plot)')
plt.show()
```

### 🔍 Interpretation:
- **Slope of the line**: Indicates strength and direction of linear relationship.
- **Shaded band**: 95% confidence interval of regression.
- **Use with hue** to detect differences in linear trends between groups.


## 4. Box Plot

```python
sns.boxplot(x='class', y='fare', data=df_titanic)
plt.title('Titanic: Fare by Class (Box Plot)')
plt.show()
```

### 🔍 Interpretation:
- **Line in the box**: Median of the group.
- **Box range**: Middle 50% (IQR).
- **Whiskers and dots**: Range and outliers.
- Useful for comparing spread and center between categories.


## 5. Violin Plot

```python
sns.violinplot(x='sex', y='age', data=df_titanic)
plt.title('Titanic: Age by Sex (Violin Plot)')
plt.show()
```

### 🔍 Interpretation:
- Combines a boxplot with a KDE.
- **Wider sections**: Higher density of values.
- **Symmetry/asymmetry**: Indicates distribution skewness.
- Great for spotting multimodal distributions.


## 6. Strip Plot

```python
sns.stripplot(x='sex', y='age', data=df_titanic, jitter=True, alpha=0.6)
plt.title('Titanic: Age by Sex (Strip Plot)')
plt.show()
```

### 🔍 Interpretation:
- Displays all data points in a group.
- **Clustered dots**: Frequent values.
- Use `jitter=True` to reduce overplotting.
- Best for small datasets or visualizing raw values.


## 7. Swarm Plot

```python
sns.swarmplot(x='sex', y='age', data=df_titanic)
plt.title('Titanic: Age by Sex (Swarm Plot)')
plt.show()
```

### 🔍 Interpretation:
- Like strip plot, but avoids overlapping dots.
- **Stacked dots**: Reflect common values.
- More readable when data are dense.
- Slower with large datasets but very informative.


## 8. Grouped Bar Plot

```python
group_counts = df_titanic.groupby(['class', 'sex']).size().unstack()
group_counts.plot(kind='bar')
plt.title('Titanic: Passenger Class by Sex (Grouped Bar)')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

### 🔍 Interpretation:
- Shows absolute counts for combinations of two categorical variables.
- Compare bars **within** and **across** groups.
- Height differences highlight associations or imbalance.


## 9. Heatmap of Crosstab

```python
import pandas as pd

ct = pd.crosstab(df_titanic['class'], df_titanic['embarked'], normalize='index')
sns.heatmap(ct, annot=True, cmap='Blues')
plt.title('Titanic: Embarkation by Class (Heatmap)')
plt.show()
```

### 🔍 Interpretation:
- **Darker colors** = higher proportion/frequency.
- Row normalization helps compare proportions within categories.
- Excellent for spotting distributional differences across groups.


## 10. Hexbin Plot

```python
import numpy as np
import pandas as pd
df_sim = pd.DataFrame({
    'age': np.random.randint(18, 80, 1000),
    'income': np.random.normal(50000, 15000, 1000)
})
plt.hexbin(df_sim['age'], df_sim['income'], gridsize=30, cmap='viridis')
plt.colorbar(label='Count')
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Simulated: Age vs. Income (Hexbin Plot)')
plt.show()
```

### 🔍 Interpretation:
- **Hex color intensity** = number of observations in that bin.
- Reveals patterns in dense scatterplots.
- Useful for large datasets where points overlap.
