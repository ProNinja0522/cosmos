# Dimensionality Reduction on Wine Dataset

This notebook applies six dimensionality reduction techniques to the `wine` dataset, which contains chemical analysis data on wines derived from three different cultivars.

Each sample has 13 numerical features. We use the following methods to project the data to 2D:
1. PCA
2. MDS
3. Isomap
4. Laplacian Eigenmap
5. t-SNE
6. Autoencoder

```python
from sklearn.datasets import load_wine
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load dataset
wine = load_wine()
X = wine.data
y = wine.target
feature_names = wine.feature_names

print(f'Data shape: {X.shape}')
print(f'Feature names: {feature_names}')
```

```python
from sklearn.preprocessing import StandardScaler
X_scaled = StandardScaler().fit_transform(X)
```

## PCA

This method reduces the 13-dimensional wine data to 2D.

```python
from sklearn.decomposition import PCA
X_pca = PCA(n_components=2).fit_transform(X_scaled)
```

```python
plt.figure(figsize=(6, 5))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='Set1', s=30, edgecolor='k')
plt.colorbar(label='Wine Class')
plt.title('PCA Projection of Wine Data')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.grid(True)
plt.show()
```

## MDS

This method reduces the 13-dimensional wine data to 2D.

```python
from sklearn.manifold import MDS
X_mds = MDS(n_components=2, random_state=42).fit_transform(X_scaled)
```

```python
plt.figure(figsize=(6, 5))
plt.scatter(X_mds[:, 0], X_mds[:, 1], c=y, cmap='Set1', s=30, edgecolor='k')
plt.colorbar(label='Wine Class')
plt.title('MDS Projection of Wine Data')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.grid(True)
plt.show()
```

## Isomap

This method reduces the 13-dimensional wine data to 2D.

```python
from sklearn.manifold import Isomap
X_iso = Isomap(n_components=2).fit_transform(X_scaled)
```

```python
plt.figure(figsize=(6, 5))
plt.scatter(X_iso[:, 0], X_iso[:, 1], c=y, cmap='Set1', s=30, edgecolor='k')
plt.colorbar(label='Wine Class')
plt.title('Isomap Projection of Wine Data')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.grid(True)
plt.show()
```

## Laplacian Eigenmap

This method reduces the 13-dimensional wine data to 2D.

```python
from sklearn.manifold import SpectralEmbedding
X_lap = SpectralEmbedding(n_components=2, random_state=42).fit_transform(X_scaled)
```

```python
plt.figure(figsize=(6, 5))
plt.scatter(X_lap[:, 0], X_lap[:, 1], c=y, cmap='Set1', s=30, edgecolor='k')
plt.colorbar(label='Wine Class')
plt.title('Laplacian Eigenmap Projection of Wine Data')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.grid(True)
plt.show()
```

## t-SNE

This method reduces the 13-dimensional wine data to 2D.

```python
from sklearn.manifold import TSNE
X_tsne = TSNE(n_components=2, random_state=42).fit_transform(X_scaled)
```

```python
plt.figure(figsize=(6, 5))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='Set1', s=30, edgecolor='k')
plt.colorbar(label='Wine Class')
plt.title('t-SNE Projection of Wine Data')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.grid(True)
plt.show()
```

## Autoencoder

This method reduces the 13-dimensional wine data to 2D.

```python
from sklearn.neural_network import MLPRegressor
ae = MLPRegressor(hidden_layer_sizes=(8,), max_iter=2000, random_state=42)
ae.fit(X_scaled, X_scaled)
X_auto = ae.predict(X_scaled)[:, :2]
```

```python
plt.figure(figsize=(6, 5))
plt.scatter(X_auto[:, 0], X_auto[:, 1], c=y, cmap='Set1', s=30, edgecolor='k')
plt.colorbar(label='Wine Class')
plt.title('Autoencoder Projection of Wine Data')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.grid(True)
plt.show()
```
