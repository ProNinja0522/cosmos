## Activity 1: Titanic Dataset â€” Distribution Plots

```python
%pip install squarify
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import squarify

# Load Titanic dataset
df_titanic = sns.load_dataset('titanic')

# Histogram of age
sns.histplot(df_titanic['age'].dropna(), bins=25, kde=False)
plt.title('Histogram of Age')
plt.show()

# KDE of fare
sns.kdeplot(df_titanic['fare'].dropna())
plt.title('KDE of Fare')
plt.show()

# Boxplot of age
sns.boxplot(x=df_titanic['age'])
plt.title('Boxplot of Age')
plt.show()

#sns.kdeplot(df_titanic['fare'].dropna())
#sns.rugplot(df_titanic['fare'].dropna(), height=0.05)
#plt.title('KDE + Rug Plot of Fare')
#plt.xlabel('Fare')
#plt.ylabel('Density')
#plt.show()

# Violin plot of age
sns.violinplot(x=df_titanic['age'])
plt.title('Violin Plot of Age')
plt.show()

# ECDF plot of fare
import numpy as np

fare = df_titanic['fare'].dropna()
x = np.sort(fare)
y = np.arange(1, len(x)+1) / len(x)
plt.plot(x, y, marker='.', linestyle='none')
plt.title('ECDF of Fare')
plt.xlabel('Fare')
plt.ylabel('ECDF')
plt.grid(True)
plt.show()

# Bar chart of class
sns.countplot(x='class', data=df_titanic)
plt.title('Bar Chart of Passenger Class')
plt.show()

# Bar chart of sex
sns.countplot(x='sex', data=df_titanic)
plt.title('Bar Chart of Sex')
plt.show()

# pie plot
counts = df_titanic['embarked'].value_counts(dropna=False)

# Plot pie chart
#counts.plot.pie(autopct='%1.1f%%', startangle=90, shadow=True)
#plt.title('Pie Chart of Embarked Port')
#plt.ylabel('')
#plt.show()


# Tree plot
counts = df_titanic['embarked'].value_counts()
labels = [f'{label}\n{count}' for label, count in zip(counts.index, counts)]

# Plot treemap
squarify.plot(sizes=counts.values, label=labels, alpha=0.8)
plt.title('Treemap of Embarked Port')
plt.axis('off')
plt.show()
```

## Activity 2: Simulated Dataset â€” Distribution Plots

```python
import numpy as np

# Simulate dataset
np.random.seed(0)
df_sim = pd.DataFrame({
    'age': np.random.randint(18, 80, 100),
    'income': np.random.normal(50000, 15000, 100),
    'gender': np.random.choice(['male', 'female'], size=100),
    'purchased': np.random.choice([0, 1], size=100)
})

# Histogram and KDE of income
sns.histplot(df_sim['income'], bins=10, kde=True)
plt.title('Histogram and KDE of Income')
plt.show()

# Boxplot of age
sns.boxplot(x=df_sim['age'])
plt.title('Boxplot of Age')
plt.show()



# Violin plot of income
sns.violinplot(x=df_sim['income'])
plt.title('Violin Plot of Income')
plt.show()

# ECDF of income
income = df_sim['income']
x = np.sort(income)
y = np.arange(1, len(x)+1) / len(x)
plt.plot(x, y, marker='.', linestyle='none')
plt.title('ECDF of Income')
plt.xlabel('Income')
plt.ylabel('ECDF')
plt.grid(True)
plt.show()

# Bar chart for gender
sns.countplot(x='gender', data=df_sim)
plt.title('Bar Chart of Gender')
plt.show()

# Bar chart for purchased
sns.countplot(x='purchased', data=df_sim)
plt.title('Bar Chart of Purchased')
plt.show()
```
