# Classical Multivariate Visualization: Iris Dataset

This notebook demonstrates five classical multivariate visualization techniques using the Iris dataset.

**Target variable:** `target` (iris species)

### Methods covered:
1. Enhanced Scatterplot
2. Pairplot with hue
3. FacetGrid / Catplot (binned sepal length)
4. Parallel Coordinates
5. (Optional) 3D Scatterplot

## 1. Enhanced Scatterplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

# Load iris dataset and create DataFrame
iris = load_iris(as_frame=True)
df_iris = iris.frame
df_iris['target'] = df_iris['target'].map(dict(enumerate(iris.target_names)))

sns.scatterplot(data=df_iris, x='petal length (cm)', y='petal width (cm)',
                hue='target', style='target', size='sepal length (cm)')
plt.title('Iris: Petal Length vs Petal Width (Colored by Species, Sized by Sepal Length)')
plt.show()
```

## 2. Pairplot with Hue

```python
sns.pairplot(df_iris, hue='target')
plt.suptitle('Iris Dataset: Pairplot with Hue', y=1.02)
plt.show()
```

## 3. Faceted Boxplot by Sepal Length Bin and Target

```python
df_iris['sepal_length_bin'] = pd.cut(df_iris['sepal length (cm)'], bins=3, labels=['Short', 'Medium', 'Long'])
sns.catplot(data=df_iris, x='sepal_length_bin', y='petal length (cm)', hue='target', kind='box')
plt.title('Iris: Petal Length by Sepal Length Bin and Species')
plt.show()
```

## 4. Parallel Coordinates

```python
from pandas.plotting import parallel_coordinates

subset = df_iris[['sepal length (cm)', 'sepal width (cm)',
                  'petal length (cm)', 'petal width (cm)', 'target']]
parallel_coordinates(subset, class_column='target', color=('#1f77b4', '#ff7f0e', '#2ca02c'))
plt.title('Iris Dataset: Parallel Coordinates')
plt.show()
```
