# Clustering and Segmentation: Iris Dataset and Gaussian Mixture Model

This notebook demonstrates five clustering methods applied to two datasets:
- **Iris dataset** (real-world)
- **Simulated Gaussian Mixture** (synthetic)

### Clustering Methods:
1. K-Means
2. Agglomerative Clustering
3. DBSCAN
4. Spectral Clustering
5. Gaussian Mixture Models (GMM)

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import adjusted_rand_score
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN, SpectralClustering
from sklearn.mixture import GaussianMixture

def plot_clusters(X, labels, title):
    plt.figure(figsize=(5, 4))
    plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='tab10', s=30, edgecolor='k')
    plt.title(title)
    plt.xlabel('Component 1')
    plt.ylabel('Component 2')
    plt.grid(True)
    plt.show()
```

## ðŸŸ¢ Iris Dataset

```python
# Load Iris dataset
iris = datasets.load_iris()
X_iris = StandardScaler().fit_transform(iris.data)
y_iris = iris.target

# Reduce to 2D with PCA for visualization
X_iris_pca = PCA(n_components=2).fit_transform(X_iris)
```

## ðŸ”µ Simulated Gaussian Mixture Dataset

```python
# Generate synthetic Gaussian mixture data
X_blob, y_blob = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=42)
X_blob = StandardScaler().fit_transform(X_blob)
```

### ðŸš€ Clustering on Iris Dataset

```python
X = X_iris_pca
y_true = y_iris
n_clusters = len(np.unique(y_true))
labels = KMeans(n_clusters=n_clusters, random_state=42).fit_predict(X)
score = adjusted_rand_score(y_true, labels)
print('Adjusted Rand Index:', round(score, 3))
plot_clusters(X, labels, 'KMeans Clustering on Iris')
```

```python
X = X_iris_pca
y_true = y_iris
n_clusters = len(np.unique(y_true))
labels = AgglomerativeClustering(n_clusters=n_clusters).fit_predict(X)
score = adjusted_rand_score(y_true, labels)
print('Adjusted Rand Index:', round(score, 3))
plot_clusters(X, labels, 'Agglomerative Clustering on Iris')
```

```python
X = X_iris_pca
y_true = y_iris
n_clusters = len(np.unique(y_true))
labels = DBSCAN(eps=0.9).fit_predict(X)
plot_clusters(X, labels, 'DBSCAN Clustering on Iris')
```

```python
X = X_iris_pca
y_true = y_iris
n_clusters = len(np.unique(y_true))
labels = SpectralClustering(n_clusters=n_clusters, random_state=42, affinity='nearest_neighbors').fit_predict(X)
score = adjusted_rand_score(y_true, labels)
print('Adjusted Rand Index:', round(score, 3))
plot_clusters(X, labels, 'Spectral Clustering on Iris')
```

```python
X = X_iris_pca
y_true = y_iris
n_clusters = len(np.unique(y_true))
labels = GaussianMixture(n_components=n_clusters, random_state=42).fit(X).predict(X)
score = adjusted_rand_score(y_true, labels)
print('Adjusted Rand Index:', round(score, 3))
plot_clusters(X, labels, 'GMM Clustering on Iris')
```

### ðŸš€ Clustering on Gaussian Mixture Dataset

```python
X = X_blob
y_true = y_blob
n_clusters = len(np.unique(y_true))
labels = KMeans(n_clusters=n_clusters, random_state=42).fit_predict(X)
score = adjusted_rand_score(y_true, labels)
print('Adjusted Rand Index:', round(score, 3))
plot_clusters(X, labels, 'KMeans Clustering on Gaussian Mixture')
```

```python
X = X_blob
y_true = y_blob
n_clusters = len(np.unique(y_true))
labels = AgglomerativeClustering(n_clusters=n_clusters).fit_predict(X)
score = adjusted_rand_score(y_true, labels)
print('Adjusted Rand Index:', round(score, 3))
plot_clusters(X, labels, 'Agglomerative Clustering on Gaussian Mixture')
```

```python
X = X_blob
y_true = y_blob
n_clusters = len(np.unique(y_true))
labels = DBSCAN(eps=0.9).fit_predict(X)
plot_clusters(X, labels, 'DBSCAN Clustering on Gaussian Mixture')
```

```python
X = X_blob
y_true = y_blob
n_clusters = len(np.unique(y_true))
labels = SpectralClustering(n_clusters=n_clusters, random_state=42, affinity='nearest_neighbors').fit_predict(X)
score = adjusted_rand_score(y_true, labels)
print('Adjusted Rand Index:', round(score, 3))
plot_clusters(X, labels, 'Spectral Clustering on Gaussian Mixture')
```
