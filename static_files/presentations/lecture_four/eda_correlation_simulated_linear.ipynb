# Simulated Dataset with Strong Linear Relationships

This notebook explores a simulated dataset containing strong linear relationships among numeric variables. It includes:
- Correlation heatmap
- Pairplot
- Clustermap

## 1. Generate Simulated Data

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Simulate data
np.random.seed(42)
n = 300
x1 = np.random.normal(50, 10, n)
x2 = 2.5 * x1 + np.random.normal(0, 5, n)
x3 = -1.2 * x1 + np.random.normal(0, 5, n)
x4 = 0.5 * x2 + 0.3 * x3 + np.random.normal(0, 3, n)
x5 = np.random.normal(0, 1, n)

df_linear = pd.DataFrame({
    'x1': x1,
    'x2': x2,
    'x3': x3,
    'x4': x4,
    'x5': x5
})

df_linear.head()
```

## 2. Correlation Heatmap

```python
corr_matrix = df_linear.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix (Simulated Data)')
plt.show()
```

## 3. Pairplot

```python
sns.pairplot(df_linear)
plt.suptitle('Pairplot of Simulated Variables', y=1.02)
plt.show()
```

## 4. Clustermap

```python
sns.clustermap(corr_matrix, annot=True, cmap='vlag')
plt.suptitle('Clustermap of Correlation Structure', y=1.05)
plt.show()
```
