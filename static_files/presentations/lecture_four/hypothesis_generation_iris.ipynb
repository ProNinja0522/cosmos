# ðŸ§  Hypothesis Generation: Iris Dataset

In this notebook, we use the **Iris dataset** to demonstrate how to generate testable hypotheses from EDA.

We'll:
1. Explore distributions and relationships
2. Identify interesting patterns
3. Formulate hypotheses that can be tested statistically later

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
df.head()
```

## ðŸ“Š Step 1: Explore Feature Distributions by Species

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='species', y='petal length (cm)')
plt.title('Petal Length by Species')
plt.grid(True)
plt.show()
```

```python
sns.pairplot(df, hue='species')
plt.suptitle('Pairwise Feature Relationships', y=1.02)
plt.show()
```

## ðŸ’¡ Step 2: Observations and Generated Hypotheses

**From the plots above, we observe:**
- Petal length appears to vary **significantly** across species.
- Setosa is clearly separable from other species in several dimensions.
- There may be overlap between versicolor and virginica on some features.

**Example hypotheses:**
1. *Mean petal length differs significantly across the three species.*
2. *Sepal width does not significantly differ between versicolor and virginica.*
3. *Setosa can be linearly separated from the other species using petal length and width.*

These hypotheses can later be tested using ANOVA, t-tests, or classification models.
